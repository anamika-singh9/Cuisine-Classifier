Data Encoding and feature engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
x=tfidf.fit_transform(df['Cuisines'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(df['Restaurant Name'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2)
```

creating functions for finding errors. 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

creating functions for finding errors. 

```python
def modelresults(predictions):
    print("Mean abslute error on model is {}".format(mean_absolute_error(Y_test,predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(Y_test,predictions))))
```

Creating models 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rfrmodel = RandomForestClassifier()
```

```python
param_gridrfr = {"max_depth" :[50,100], "n_estimators" : [2,5,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, Y_train)
```

```python
randomforestpredictions = gridrfr.predict(X_test)
```

```python
modelresults(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
import joblib
joblib.dump(gridrfr,"mlmodel.pkl")
joblib.dump(le,"le.pkl")
joblib.dump(tfidf,"tfidf.pkl")
```
